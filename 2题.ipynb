```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import sklearn
```

```python
data=pd.read_csv("./bb.csv")
print(data)
```

```python
X=np.c_[data['行驶里程（1000英里）']]
y=np.c_[data['价格（1000美元）']]
```

```python
data.plot(kind="scatter",x="行驶里程（1000英里）",y="价格（1000美元）")
plt.show()
```

```python
from sklearn import linear_model
lr_model=linear_model.LinearRegression()
lr_model.fit(X,y)
print("斜率:%s,截距:%s" %(lr_model.coef_[0][0],lr_model.intercept_[0]))
print("估计模型为:y=%sx+%sy"%(lr_model.coef_[0][0],lr_model.intercept_[0]))
```

```python
data.plot(kind="scatter",x="行驶里程（1000英里）",y="价格（1000美元）")
plt.plot(X,lr_model.predict(X.reshape(-1,1)),color='red',linewidth=4)
plt.show()
```
